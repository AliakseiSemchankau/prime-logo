Предварительно мы генерируем простые числа первого миллиона, чтобы проверять делимость n сперва на малые простые числа.

Делимость на эти числа оставляет приблизительно $1/\log{1000000} \approx 7.2\%$ от всех натуральных чисел. 

```python
def brute_prime(n):
  x = 2
  while x * x <= n:
    if n % x == 0:
      return False
    x += 1
  return True

primes = [p for p in range(2, 1000000) if brute_prime(p)]
print('# of primes: {}'.format(len(primes)))
print(primes[:20])
```

Функция pow(a, b, p) возвращает $a^b \mod p$.

Функция generate_random(n) конструирует случайное число из $2\ln{n}$ бит и возвращает его по модулю $n$.

Функция is_prime(n, rounds=10, randomize=False) возвращает True если (достаточно большое) число $n$ простое и False иначе. 
Сперва проверяется делимость $n$ на простые, меньшие миллиона. Затем генерируются числа $a$ (если рандомизация выключена, то $a = 2, 3, \dots$) и с помощью вероятностного теста Миллера-Рабина проверяется, является ли $a$ свидетелм простоты числа $n$.

Если алгоритм возвращает False, то число гарантировано составное. Если алгоритм возвращает True, то число простое с вероятностью $1 - 4^{-r}$, где $r$=rounds. 

```python
from random import randint
import math

def pow(a, b, p):
  res = 1
  x = a
  while b > 0:
    if b % 2 == 1:
      res = (res * x) % p
    b //= 2
    x = (x * x) % p
  return res

def generate_random(n):
  digs = round(math.log(n))
  res = 0
  pow2 = 1
  for i in range(1, 2 * digs): # this implies res to have 1.4 * #digits of n
    dig = randint(0, 1)
    res = (res + dig * pow2) % n
    pow2 = (2 * pow2) % n
  return res

def is_prime(n, rounds = 10, randomize=False, print_round=False):
  for p in primes:
    if n % p == 0:
      return False

  d = n - 1
  s = 0
  while d % 2 == 0:
    s += 1
    d //= 2

  def miller_rabin_round(a):
    x = pow(a, d, n)
    if x == 1 or x == n - 1:
      return False
    # from 1 to s-1
    for j in range(1, s):
      x = (x * x) % n
      if x == n - 1:
        return False
    return True

  # one round takes ~7 secs for 4000-digit n 
  # experiment have shown 119 rounds to take 13 mins 45 secs
  for i in range(rounds):
    if print_round:
      print('round={}'.format(i+1))
    if not randomize:
      a = i+2
    else:
      a = generate_random(n)
    if miller_rabin_round(a):
      if i > 0:
        print('it took {} rounds'.format(i+1))
      return False
  return True
```

```python
def ascii_to_n(ascii):
  return int(''.join(ascii))
```

В большом числе, заданном изображением, мы пробуем заменить одну из цифр $7, 0, 2$ на одну из цифр $3, 5, 6, 8, 9$ и получить простое число. Это удается сделать после приблизительно $2500$ попыток.

Вероятность того, что число из $5000$ знаков является простым, равняется приблизительно $1/\ln{10^{5000}} = 1/(5000\ln{10})$. By design, мы рассматриваем только нечетные числа, что удваивает вероятность, и превращает ее в $1/(2500\ln{10}) \approx 1/5700$. Значит, среди $5700$ произвольных нечетных чисел из $5000$ цифр в среднем есть одно простое число. Таким образом, матожидание числа попыток равно $5700$, но нам повезло найти простое число за $2500$ попыток.

```python
def try_replace(digits, i, j, new_digit):
      old_str = digits[i]
      new_str = list(digits[i])
      new_str[j] = new_digit
      digits[i] = ''.join(new_str)
      n = ascii_to_n(digits)
      new_digits = None
      flag = False
      if is_prime(n):
        flag = True
        new_digits = digits.copy()
      digits[i] = old_str
      return flag, new_digits

# here we try to replace some orange/blue/azure pixel in array 'digits' 
# with digit 3/5/6/8/9 and check if resulted ascii represents a prime number
# if so, it is stored in array 'new_digits'
# we also print every 20th iteration
colored = [orange, blue, azure]
pointy_boi = ['3', '5', '6', '8', '9']
flag = False
iteration = 0
for i in range(len(digits)):
  if flag:
    break
  for j in range(len(digits[0])):
    if flag:
      break
    if not digits[i][j] in colored:
      continue
    for new_digit in pointy_boi:
      if flag:
        break
      iteration += 1
      if iteration < 2500:
        continue
      if iteration % 20 == 0:
        print('iteration=', iteration)
      result, new_digits = try_replace(digits, i, j, new_digit)
      if result:
        print('success! changes [{}, {}] from {} to {}'.format(i, j, digits[i][j], new_digit))
        flag = True

if flag:
  for line in new_digits:
    print(line)
```

Генерируем .html-страницу, где каждая цифра имеет белый цвет и фон одного из пяти цветов -- серого, черного, оранжевого, синего или бирюзового.

Шрифт выбран Courier (является моноширинным).

Полученную выше пятерку мы также красим в синий цвет.

```python
ch_to_color = {
    white: 'gray',
    black: 'black',
    orange: 'F58C29',
    blue: '0074C8',
    azure: '1ECECA',
    '5': '0074C8'
}

def ascii_to_html(ascii):
  header = '''
<html>
<code>
<span 
class="ascii" 
style=
"
-webkit-print-color-adjust: exact;
color: white; 
background: white;
display:inline-block;
white-space:pre;
letter-spacing:0;
line-height:1.13;
font-family:Courier;
font-size:14px;
">
'''

  lines = []

  for ascii_line in ascii:
    line = ''
    for ch in ascii_line:
      color = ch_to_color[ch]
      line += '<span style="background-color:{};">{}</span>'.format(color, ch)
    lines.append(line)
  
  ending = '''
    </span>
    </code>
    </html>
'''

  return '\n'.join([header, *lines, ending])

```
